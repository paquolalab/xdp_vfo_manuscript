```python
import pandas as pd
from collections import Counter
import session_info
```

```python
#format the dataframe to remove spaces and/or conflicting characters
df = pd.read_excel('../_h/XDP striatal organoids RNAseq_sample list_LD_JE_June2021.xlsx',skiprows=2)
df = df.rename(columns={"Unnamed: 0": "Family_Group", 'iPSC line': 'iPSC_line', 'Cell line type':'Cell_line','Day of maturation': 'Day_of_Maturation',
                       'Kit for RNA extraction':'Kit','disease variable': 'Disease_Status'})
df['Differentiation_cohort'] = df['Differentiation_cohort'].fillna(method='ffill')

df['Differentiation_cohort'] = df['Differentiation_cohort'].str.replace(' ','_',regex = True)
df['Differentiation_cohort'] = df['Differentiation_cohort'].str.replace('_$|__$','',regex = True)
df['Cell_line'] = df['Cell_line'].str.replace(' - ','-',regex = True)
df['Kit'] = df['Kit'].str.replace(', #','_',regex = True)


##change the delta symbol
df['Cell_line'] = df['Cell_line'].str.replace('\u0394',"DELTA-",regex = True)
df.drop('Unnamed: 10',axis=1,inplace=True)

df.loc[9]=df.loc[52] #remove sample 10N and add 53R inplace
df.at[9,'Differentiation_cohort'] = 'A' # Change 53R differentiation_cohort to group A
df = df[:-1] #drop last row
df.Sample = df.Sample.astype(str)
#df
```

```python
qc = pd.read_csv('../../../speaqeasy_star/_m/count_objects/read_and_alignment_metrics_Jlab_experiment.csv')
qc = qc.drop(['Unnamed: 0'], axis=1)
qc.SAMPLE_ID = qc.SAMPLE_ID.astype(str)
```
